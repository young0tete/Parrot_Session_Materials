# 분포 시각화

```python
import numpy as np 
import pandas as pd 
from scipy import stats 
import matplotlib.pyplot as plt 
import seaborn as sns     # 시각화
from sklearn import preprocessing
import warnings
warnings.filterwarnings( 'ignore' )
```

## 정규분포

```python
legend = []
x = np.linspace(-5, 5, 101)   # x 정의

def norm (mu, var) :
    legend.append("N(" + str(mu) + ", " + str(var) + ")")
    return stats.norm(mu, var).pdf(x)        

plt.figure(figsize=(12, 8))          

for i in np.arange(4) + 1:
    plt.plot(x, norm(0, i))         
    
plt.xlabel("x")                     
plt.ylabel("y")                     
plt.grid()                          
plt.title(r'$Normal Distribution$')     
plt.legend(legend)                   
plt.show()                           
```

## 표준정규분포

```python
x = np.linspace(-5, 5, 101)   # x 정의
y = stats.norm(0, 1).pdf(x)        

def norm (mu, var) :
    return stats.norm(mu, var).pdf(x)  

plt.figure(figsize=(10, 6))          
plt.plot(x, norm(0,1))         
plt.xlabel("x")                      
plt.ylabel("y")                                 
plt.title('Standard Normal Distribution')
plt.legend(["Standard Normal Distribution"])              
plt.show()           
```

## t-분포

```python
x = np.linspace(-5, 5, 101)
y = stats.t(10).pdf(x)

plt.figure(figsize=(10,6))
plt.plot(x, y, color="green")         
plt.xlabel("t")                      
plt.ylabel("y")                                             
plt.title("t Distribution")     
plt.legend(["t with 10 degrees of freedom"])   
plt.show()                           
```

## F-분포

```python
import scipy.special as sp
import math
x = np.linspace(0, 5, 201)

v1 = 4
v2 = 50

y = stats.f(v1, v2).pdf(x)        

plt.figure(figsize=(10, 6))
plt.plot(x, y, label='F(4, 50)')
plt.xlabel('F')
plt.ylabel("y")
plt.grid()     
plt.title('F Distribution')
plt.legend()
plt.show()
```

## 이항분포

```python
N = 30
mu = 0.5
y = stats.binom(N, mu)

x = np.arange(N + 1)
plt.plot(x, y.pmf(x),'-')
plt.xlabel("sampled value")
plt.ylabel("P(x)")
plt.title('Binomial Distribution')
plt.show()
```

## 카이스퀘어분포

```python
df = 2          # 자유도
x = np.linspace(0, 20, 201)
y = stats.chi2(df).pdf(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b')
plt.xlabel('X')
plt.ylabel('P(X)')
plt.grid()
plt.title(r'$\chi^2$ Distribution (df = 2)')
plt.show()
```

## 포아송 분포

```python
Lambda = 20
Y = stats.poisson(Lambda)
X = np.arange(41, dtype='f8')

plt.plot(X, Y.pmf(X))
plt.xlabel('sampled value')
plt.ylabel('P(x)')
plt.title(r'$Poisson Distribution(\lambda = 20)$')
plt.show()
```

```python
plt.figure(figsize=(12, 8))

X = np.arange(41, dtype='f8')
for l in range(4, 30, 4):
    plt.plot(X, stats.poisson(l).pmf(X),'o--',
             label=r'$(\lambda =$' + str(l) + ')')
    
plt.xlabel('X')
plt.ylabel('P(X)')
plt.title(r'$Poisson Distribution$')
plt.grid()    # 격자
plt.legend()    # 범례 표시
plt.show()
```

## + 베르누이 분포

```python
mu = 0.5
y = stats.bernoulli(mu)    # 베르누이 함수

x = [0, 1]
plt.bar(x, y.pmf(x))    # y.pmf(xx) = array([0.5, 0.5])
plt.xlim(-1, 2)
plt.ylim(0, 1)
plt.xticks([0, 1], ["x=0", "x=1"])
plt.xlabel("sampled value")
plt.ylabel("P(x)")
plt.title('Bernoulli Distribution')
plt.show()
```
